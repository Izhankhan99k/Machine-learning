```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier,plot_tree
from sklearn.model_selection import train_test_split
import dtreeviz
```

```python
iris=load_iris()
x=iris.data
y=iris.target
dtc=DecisionTreeClassifier(max_depth=4)
dtc.fit(x,y)
df=pd.DataFrame(x,y)
x
```

```python
plot_tree(dtc)
```

```python
viz=dtreeviz.model(dtc,x,y,feature_names=iris.feature_names,class_names=["setosa","versicolor","virginica"])
viz.view()

```

```python
dtc.predict(np.reshape([0.5,1.2,5.5,1],(1,-1)))
```
